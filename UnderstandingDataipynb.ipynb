Sample Selection:

Demographic Factors:

Gender:

 Biases can emerge if there is an underrepresentation or overrepresentation of one gender in the dataset.

Age:

 The dataset may not accurately reflect different age groups, leading to age-related bias.
Race and Ethnicity: Bias can occur if certain racial or ethnic groups are underrepresented or if data collection methods are not culturally sensitive.
Socioeconomic Status:

Income Disparities:

 Biases may be present if the dataset disproportionately represents individuals from certain income levels, leading to skewed results.
Geographic Location:

Regional Bias:

If data is collected predominantly from specific geographic regions, it may not generalize well to other areas.
Education Level:

Educational Bias:

If the dataset is biased towards individuals with certain levels of education, it may not capture the experiences of those with different educational backgrounds.
Cultural and Linguistic Factors:

Language Bias:

 If data is collected primarily in one language, it may not be inclusive of speakers of other languages.

Cultural Norms:

Biases can arise if data collection methods are influenced by specific cultural norms.
Historical Biases:

Legacy Biases:

Historical biases present in past data collection practices can persist and impact current datasets.
Measurement and Instrumentation:

Measurement Bias:

If the instruments used for data collection are biased, it can lead to inaccuracies.
Observer Bias: Bias may arise if the individuals collecting or interpreting the data have preconceived notions or perspectives.
Algorithmic Biases:

Training Data Bias:

In machine learning, biases can be introduced if algorithms are trained on biased datasets.

```python
import pandas as pd
from imblearn.under_sampling import RandomUnderSampler

# Load the dataset
data = pd.read_csv('recividism.csv')  # Assuming 'recidivism.csv' is the name of your dataset

# Before undersampling
print("Before undersampling:")
print(data['sex'].value_counts())

# Define features (X) and target variable (y)
X = data.drop(columns=['sex'])
y = data['sex']

# Define the undersampling strategy
undersampler = RandomUnderSampler(random_state=42)

# Apply undersampling to balance the dataset
X_resampled, y_resampled = undersampler.fit_resample(X, y)

# After undersampling
print("\nAfter undersampling:")
print(pd.Series(y_resampled).value_counts())
```

```python
# Calculate the proportions before and after undersampling
before_proportions = data['sex'].value_counts(normalize=True)
after_proportions = pd.Series(y_resampled).value_counts(normalize=True)

# Calculate the bias reduction for each class
bias_reduction = before_proportions - after_proportions

print("Bias reduction after undersampling:")
print(bias_reduction)
```


The bias reduction after undersampling shows a decrease in bias towards the majority class (Male) and an increase in bias towards the minority class (Female). In this case:

Bias reduction for Female: -0.315298
Bias reduction for Male: 0.315298
A negative value for Female indicates a decrease in bias towards this class, while a positive value for Male indicates an increase in bias towards this class. This suggests that the undersampling process has helped to balance the dataset by reducing the bias towards the majority class (Male) and increasing the representation of the minority class (Female).

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
from imblearn.over_sampling import SMOTE

# Load the dataset
data = pd.read_csv("recividism.csv")

# Select only the columns of interest
columns_of_interest = ['race', 'sex', 'is_violent_recid', 'juv_other_count']
data_subset = data[columns_of_interest]

# Convert non-numeric columns to categorical data type
categorical_columns = data_subset.select_dtypes(include=['object']).columns
data_subset[categorical_columns] = data_subset[categorical_columns].astype('category')

# Encode categorical variables
label_encoders = {}
for col in categorical_columns:
    label_encoders[col] = LabelEncoder()
    data_subset[col] = label_encoders[col].fit_transform(data_subset[col])

# Handle missing values
imputer = SimpleImputer(strategy='most_frequent')
data_subset_imputed = pd.DataFrame(imputer.fit_transform(data_subset), columns=data_subset.columns)

# Perform SMOTE for oversampling minority classes
X = data_subset_imputed.drop(columns=['sex'])  # Features
y = data_subset_imputed['sex']  # Target variable
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Convert the resampled data back to DataFrame
resampled_df = pd.DataFrame(X_resampled, columns=X.columns)
resampled_df['sex'] = y_resampled

# Inverse transform categorical encoding
for col in categorical_columns:
    resampled_df[col] = label_encoders[col].inverse_transform(resampled_df[col])

# Show the resampled DataFrame
print(resampled_df)
```

his code performs the following steps:

Selects the columns of interest.
Converts non-numeric columns to categorical data type.
Encodes categorical variables using LabelEncoder.
Handles missing values by imputing the most frequent value.
Performs SMOTE analysis to oversample the minority class (females) in the sex column.
Converts the resampled data back to a DataFrame and inverse transforms the categorical encoding.


```python
# Calculate the proportion of males and females before SMOTE
before_smote_proportion = data['sex'].value_counts(normalize=True)

# Calculate the proportion of males and females after SMOTE
after_smote_proportion = resampled_df['sex'].value_counts(normalize=True)

# Calculate the change in bias
bias_reduction = before_smote_proportion - after_smote_proportion

# Print the change in bias
print("Bias reduction after SMOTE:\n", bias_reduction)
```

Original Sample (X_ori): This is the data point in the minority class that we want to create synthetic samples for. In the slide, it's represented by the coordinates (4,7).

Neighbor Sample (X_neig): This is a nearby data point to X_ori, chosen from the k-nearest neighbors in the minority class. In the slide, the neighbor's coordinates are (6,8).

Factor: This is a random number between 0 and 1. It's used to determine how far along the line between the original sample and its neighbor the synthetic sample will be placed. In the slide, this factor is 0.8.

The difference between X_ori and X_neig is calculated as ((4,7) - (6,8)) which equals (-2,-1).
Then, this difference is multiplied by the factor 0.8, giving us 0.8 * (-2,-1) = (-1.6, -0.8).

Now, in the context of undersampling, SMOTE can be combined with undersampling to balance the dataset effectively. This hybrid approach typically involves oversampling the minority class using SMOTE and then using an undersampling technique to reduce the number of examples in the majority class. The steps for combining SMOTE with undersampling are as follows:

Oversample the Minority Class: Apply SMOTE to the minority class to increase its representation in the dataset.

Undersample the Majority Class: Once the minority class has been augmented, use an undersampling method (such as random undersampling, Tomek links removal, or Cluster Centroids) to reduce the size of the majority class, thereby balancing the dataset.

The undersampling step does not involve the synthetic sample generation used by SMOTE; it typically involves random deletion of instances from the majority class or more sophisticated methods that try to preserve the majority class's information while reducing its size.

Synthetic Minority Over-sampling Technique (SMOTE) is a popular method used to address class imbalance in machine learning datasets. It was introduced by Chawla et al. in 2002. SMOTE works by generating synthetic samples for the minority class, thereby balancing the class distribution.

Here's how SMOTE works:

Identify Minority Class: First, you identify the minority class in your dataset, typically the class with fewer instances.

Select Data Points: SMOTE selects individual instances from the minority class.

Find Nearest Neighbors: For each selected instance, SMOTE finds its k nearest neighbors in the feature space. These neighbors are usually determined using some distance metric, such as Euclidean distance.

Generate Synthetic Samples: SMOTE creates synthetic samples by selecting random points along the line segments joining the selected instance and its nearest neighbors.

Add Synthetic Samples: These synthetic samples are added to the dataset, effectively increasing the number of instances in the minority class.

By generating synthetic samples, SMOTE helps to increase the representation of the minority class, thereby reducing class imbalance. This can lead to better model performance, especially in scenarios where the minority class is of interest. However, it's essential to use oversampling techniques judiciously, as they can also introduce noise and overfitting if not applied carefully. Additionally, evaluating the performance of the model on balanced metrics like F1-score or precision-recall curves is crucial when dealing with imbalanced datasets.

Undersampling vs Oversampling

1. In extreme cases where the number of observations in the rare class(es) is really small, oversampling is better.

2. The main advantage of undersampling is that data scientists can correct imbalanced data to reduce the risk of their analysis or machine learning algorithm skewing toward the majority

3. Oversampling can introduce sampling error, which can lead to biased results. Additionally, oversampling can increase the risk of overfitting, where the model learns the noise in the dataset
